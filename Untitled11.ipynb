# PREPOSESSING

```python
#@title *Open File Acquisition*
import re
import numpy as np
import nltk

prep = pd.read_csv(prepInput,delimiter=',')    # membuka file acquisition
prep = prep[[prepGetColumn]]
prep
```

```python
#@title *Case Folding*
def caseFolding(text):  # function untuk menyamakan huruf dalam text
  text = text.lower()
  return text
prep = prep[[prep.columns[len(prep.columns)-1]]]
prep.rename(columns={prep.columns[len(prep.columns)-1]: prepNewColumn},inplace=True)
prep[prepListColumn[0]] = prep[prep.columns[len(prep.columns)-1]].apply(lambda x: caseFolding(x))
prep
```

```python
#@title *Cleaning*
def cleaning(text):   # function untuk membersihkan symbol dalam text
  text = re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)"," ",text)
  text = re.sub(r'^RT[\s]+', '', text)
  text = re.sub('/n', ' ',text)
  text = re.sub(r'((www\.[^\s]+)|(https?://[^\s]+)|(http?://[^\s]+))',' ',text)
  text = re.sub(r' +', ' ',text)
  return text
prep = prep[[prep.columns[len(prep.columns)-1]]]
prep.rename(columns={prep.columns[len(prep.columns)-1]: prepNewColumn},inplace=True)
prep[prepListColumn[1]] = prep[prep.columns[len(prep.columns)-1]].apply(lambda x: cleaning(x))
prep
```

```python
#@title *Word Normalization*
try:
    keyNorm
except:
    keyNorm = pd.read_csv(prepKeyNorm, sep=';', header=None)
def textNormalize(text):  # function untuk mengubah kata sesuai ejaan

    text = ' '.join([keyNorm[keyNorm[keyNorm.columns[0]] == word][keyNorm.columns[1]].values[0] if (keyNorm[keyNorm.columns[0]] == word).any() else word for word in text.split()])
    return text
prep = prep[[prep.columns[len(prep.columns)-1]]]
prep.rename(columns={prep.columns[len(prep.columns)-1]: prepNewColumn},inplace=True)
prep[prepListColumn[3]] = prep[prep.columns[len(prep.columns)-1]].apply(lambda x: textNormalize(x))
prep
```

```python
#@title *Stop Removing*
try:
  stopword = nltk.corpus.stopwords.words('indonesian')
except:
  nltk.download('stopwords')
finally:
  stopword = nltk.corpus.stopwords.words('indonesian')
def stopRemoving(text):  # function untuk menghapus stopword dalam text
    text = re.split('\W+', text)
    text = [word for word in text if word not in stopword]
    text = np.array(text)
    text = ' '.join(text)
    return text
prep = prep[[prep.columns[len(prep.columns)-1]]]
prep.rename(columns={prep.columns[len(prep.columns)-1]: prepNewColumn},inplace=True)
prep[prepListColumn[2]] = prep[prep.columns[len(prep.columns)-1]].apply(lambda x: stopRemoving(x))
prep
```

```python
#@title *Stemming*
try:
	  import Sastrawi
except ImportError:
    !pip install Sastrawi
finally:
    import Sastrawi
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
factory = StemmerFactory()
stemmer = factory.create_stemmer()
def stemming(text):  # function untuk stemming text
    return stemmer.stem(text)
prep = prep[[prep.columns[len(prep.columns)-1]]]
prep.rename(columns={prep.columns[len(prep.columns)-1]: prepNewColumn},inplace=True)
prep[prepListColumn[4]] = prep[prep.columns[len(prep.columns)-1]].apply(lambda x: stemming(x))
prep
```

```python
#@title *Tokenization*
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize
def tokenization(text):  # function untuk tokenize text
    return word_tokenize(text)
prep = prep[[prep.columns[len(prep.columns)-1]]]
prep.rename(columns={prep.columns[len(prep.columns)-1]: prepNewColumn},inplace=True)
prep[prepListColumn[5]] = prep[prep.columns[len(prep.columns)-1]].apply(lambda x: tokenization(x))
prep
```

# LEXICON MODEL

```python
lex = pd.read_csv(lexInput,delimiter=',')    # membuka file acquisition
lex[[lex.columns[len(lex.columns)-1]]]
lex.rename(columns={lex.columns[len(lex.columns)-1]: lexNewColumn},inplace=True)

lex[lex.columns[len(lex.columns)-1]] = lex[lex.columns[len(lex.columns)-1]].apply(lambda x:cleaning(x))
lex[lex.columns[len(lex.columns)-1]] = lex[lex.columns[len(lex.columns)-1]].apply(lambda x: tokenization(x))
lex
```

```python
lexicon_positive = pd.read_csv(lexPositive)
lexicon_positive_dict = {}
for index, row in lexicon_positive.iterrows():
    if row[0] not in lexicon_positive_dict:
        lexicon_positive_dict[row[0]] = row[1]

lexicon_negative = pd.read_csv(lexNegative)
lexicon_negative_dict = {}
for index, row in lexicon_negative.iterrows():
    if row[0] not in lexicon_negative_dict:
        lexicon_negative_dict[row[0]] = row[1]

def lexiconIndonesia(text):
    polarity_score = 0
    positive = ''
    negative = ''
    neutral = ''
    for word in text:
        score = 0
        if (word in lexicon_positive_dict):
            score = score + lexicon_positive_dict[word]
        if (word in lexicon_negative_dict):
            score = score + lexicon_negative_dict[word]
        if (score >= 1 ):
            positive = f"{word} {positive}"
        if (score == 0 ):
            neutral = f"{word} {neutral}"
        if (score <= -1 ):
            negative = f"{word} {negative}"
        polarity_score =  polarity_score+score
    polarity=''
    if (polarity_score > 0):
        polarity = 1
    elif (polarity_score < 0):
        polarity = -1
    else:
        polarity = 0
    return polarity_score,polarity,neutral,positive,negative
results = lex[lexNewColumn].apply(lambda x:lexiconIndonesia(x))
results = list(zip(*results))
lex[lexListColumn[0]] = results[0]
lex[lexListColumn[1]] = results[1]
lex[lexListColumn[2]] = results[2]
lex[lexListColumn[2]] = lex[lexListColumn[2]].apply(lambda x: tokenization(x))
lex[lexListColumn[3]] = results[3]
lex[lexListColumn[3]] = lex[lexListColumn[3]].apply(lambda x: tokenization(x))
lex[lexListColumn[4]] = results[4]
lex[lexListColumn[4]] = lex[lexListColumn[4]].apply(lambda x: tokenization(x))
lex.to_csv(lexOutput, index = False)  # menyimpan  Data lexicon
lex= pd.read_csv(lexOutput,delimiter=',') # checking file yang telah di simpan

label = lex[labelListColumn]
label[lexNewColumn] = label[lexNewColumn].apply(lambda x:cleaning(x))
label.to_csv(labelOutput, index = False)  # menyimpan  Data Label
label= pd.read_csv(labelOutput,delimiter=',') # checking file yang telah di simpan
label
```

# PENGUJIAN TF-IDF


```python
data= pd.read_csv(labelOutput,delimiter=',')
data[data['polarity']==1]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# create a TfidfVectorizer object
vectorizer = TfidfVectorizer()
textTF = vectorizer.fit_transform(data['lastText'])
print(textTF)
```

```python
# Splitting Datafrom sklearn.model_selection import train_test_split
from sklearn.model_selection import train_test_split

xTrain, xTest, yTrain, yTest = train_test_split(textTF, data['polarity'], test_size=0.2, random_state = 0)

print('Number transactiion x_train dataset:', xTrain.shape)
print('Number transactiion x_test dataset:', xTest.shape)
print('Number transactiion y_train dataset:', yTrain.shape)
print('Number transactiion y_test dataset:', yTest.shape)
print('Number transaction y_test dataset:', yTest.shape)
```

#NAIVES BAYES

```python
import numpy as np
from sklearn.naive_bayes import GaussianNB

GNB = GaussianNB()
GNB.fit(xTrain.toarray(), yTrain)
# Predict on test set
yPred = GNB.predict(xTest.toarray())
```

```python
from sklearn.model_selection import cross_val_score
# Melakukan validasi silang dengan Gaussian Naive Bayes
crossVal = cross_val_score(GNB, xTrain.toarray(), yTrain, cv=10, scoring='accuracy')

print('Akurasi setiap split:', crossVal, '\n')
print('Rata-rata akurasi pada cross validation:', crossVal.mean())
```

```python
from sklearn.metrics import accuracy_score
# Menghitung akurasi prediksi
scoreGNB = accuracy_score(yTest, yPred)

print("Akurasi Gaussian Naive Bayes:", scoreGNB)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, recall_score, precision_score, f1_score

print('Confusion matrix:\n', confusion_matrix(yTest, yPred))

# Menampilkan laporan klasifikasi
print('Classification report:\n', classification_report(yTest, yPred))

# Menampilkan metrik evaluasi untuk micro average
print('Micro Average:')
print("GaussianNB Recall Score -> ", recall_score(yTest, yPred, average='micro') * 100)
print("GaussianNB Precision Score -> ", precision_score(yTest, yPred, average='micro') * 100)
print("GaussianNB F1 Score -> ", f1_score(yTest, yPred, average='micro') * 100)
```

# WORDCLOUD


```python
from wordcloud import WordCloud, STOPWORDS
from collections import Counter
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
sns.set(style = 'whitegrid')

wCloud= pd.read_csv(wCloudInput,delimiter=',')
wCloud = wCloud[wCloudListColumn]
wCloud
```

```python
# word cloud untuk semua kata
setTittle = ['Dataset','Neutral','Positive','Negative']
def wordCloudShow(x):
  listWord = cleaning(' '.join([i for i in wCloud[wCloud.columns[x]]]))
  cloud = WordCloud(background_color = "white", max_words = 200, stopwords = set(STOPWORDS)).generate(listWord)
  fig, ax = plt.subplots(figsize = (5, 4))
  ax.set_title( "WordCloud "+ setTittle[x] +" KDRT", fontsize = 13)
  ax.grid(False)
  ax.imshow(cloud, interpolation='bilinear')
  fig.tight_layout(pad=0)
  ax.axis('off')
  plt.show()


  word_freq = Counter(listWord.split())
  top_words = word_freq.most_common(10)
  words = [word[0] for word in top_words]
  freqs = [word[1] for word in top_words]
  plt.figure(figsize=(5, 4))
  plt.bar(words, freqs)
  plt.xlabel('Words')
  plt.ylabel('Frequency')
  plt.title( setTittle[x]+" Words Frequency", fontsize = 13)
  plt.xticks(rotation=45)
  plt.show()
  print('')

  plt.plot(words, freqs, color='green', linestyle='dashed', linewidth = 3,
          marker='o', markerfacecolor='blue', markersize=13)

  plt.xlabel('Words')
  plt.ylabel('Frequency')
  plt.title( setTittle[x]+" Words Frequency", fontsize = 13)
  plt.xticks(rotation=45)
  plt.show()

for i in range(len(setTittle)):
  wordCloudShow(i)
```
